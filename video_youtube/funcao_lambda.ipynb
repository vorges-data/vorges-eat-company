```python
def create_price_tye(Price_range):
    if Price_range == 1:
        return "cheap"
    elif Price_range == 2:
        return "normal"
    elif Price_range == 3:
        return "expensive"
    else:
        return "gourmet"
```

```python
df['Preço_text'] = df.loc[:, 'Price range'].apply( lambda x: create_price_tye(x))
```
